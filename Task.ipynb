5. Skewness (Skew)
What is skewness?

It tells:

Is data leaning left or right?

Think of it like data shape.

1. Symmetrical (Normal)

Balanced both sides.

Skew ≈ 0

2. Positive Skew (Right Skew)

Long tail right side.

Skew > 0

Example:
Salary data usually.

Few people earn very high salary.

3. Negative Skew (Left Skew)

Long tail left side.

Skew < 0

Easy Rule
Skew Value	Meaning
0	  Normal
> 0   Right skew
< 0	  Left skew

```python
data["Current_Salary"].skew()
```

```python
data["Employee_age"].skew()
```

6. Kurtosis
What is kurtosis?

Kurtosis tells:

How many extreme values (outliers) exist?

OR

How sharp or flat data is

High kurtosis

Many outliers.

Sharp peak.

Low kurtosis

Few outliers.

Flat shape.

```python
data["Current_Salary"].kurt()
```

```python
data["years_experience"].kurt()
```

Easy Rule
Value	Meaning
> 3	    More outliers
< 3	    Fewer outliers

(For now just remember this simple rule.)

What is Correlation?

Correlation means:

Relationship between two columns

It tells:

If one thing changes, does another thing also change?

Example:

More experience → More salary?
Better performance → Higher salary?
More projects → Higher satisfaction?

This is called correlation.

Correlation Values

Correlation gives values between -1 to +1

−1≤r≤1

Value	Meaning
+1	Strong positive relation
0	No relation
-1	Strong negative relation

1. Positive Correlation (+)

If one increases, another also increases.

Example:

Experience ↑ → Salary ↑

2. Negative Correlation (-)

If one increases, another decreases.

Example:

Work stress ↑ → Satisfaction ↓

3. No Correlation (0)

No relationship.

Example:

Employee ID ↔ Salary

No connection.

Find correlation

```python
data.corr(numeric_only = True)
```

This gives correlation between all numeric columns.

You’ll see a table like this:

	             Salary	     Experience	     Rating
Salary	         1.00	     0.75	         0.30
Experience	     0.75	     1.00	         0.20

Easy Understanding:0.75

means: Strong relationship between salary and experience.

So: More experience → higher salary
