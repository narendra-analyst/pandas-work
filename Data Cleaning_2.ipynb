Correlation:

This dataset is excellent for learning correlation.

Select numerical columns:

```python
numeric_df = df.select_dtypes(include='number')

numeric_df.corr()
```

Heatmap:

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```

Data Visualization Practice
Population by Continent

```python
df.groupby('Continent')['2022 Population'].sum().plot(kind='bar')
plt.show()
```

Top 10 Countries Population

```python
top10 = df.sort_values(
    by='2022 Population',
    ascending=False
).head(10)

top10.plot(
    x='Country',
    y='2022 Population',
    kind='bar'
)
plt.show()
```

1. Which countries grew the fastest?

```python
df.sort_values(
    by='Growth Rate',
    ascending=False
).head(10)
```

2. Which countries lost population?

```python
df[df['2022 Population'] < df['2020 Population']]
```

3. What percentage of the world's population lives in each continent?

```python
continent_pop = df.groupby(
    'Continent'
)['2022 Population'].sum()

continent_pop / continent_pop.sum() * 100
```
