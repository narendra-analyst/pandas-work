# 3 types of Distribution

# 1. Histogram (histplot)

Shows the frequency distribution of a variable.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.histplot(data=tips, x="total_bill")
plt.show()
```

```python
#Uses:
Understand data distribution
Check data spread
Identify skewness
```

# 2. KDE Plot (Kernel Density Estimation) (kdeplot)

Shows a smooth curve representing the probability density of the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.kdeplot(data=tips, x="total_bill")
plt.show()
```

```python
# Uses:
Visualize the shape of the distribution
Identify peaks (modes)
Compare distributions smoothly
```

# 3. Distribution Plot (displot)

A figure-level function that can create histograms, KDE plots, and other distribution visualizations.

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.displot(data=tips, x="total_bill", kde=True)
```

```python
# Uses:
Combined histogram and KDE
Easy distribution analysis
Supports faceting by categories
```

```python
#Quick Comparison:
#     Plot	                   Function	                              Purpose
Histogram	                sns.histplot()	              Shows frequency of values
KDE Plot	                sns.kdeplot()	              Shows smooth density curve
Distribution Plot	        sns.displot()	              Overall distribution analysis
```

# Bivariate Distribution

# 1. Regplot
regplot() is a Seaborn function used to create a scatter plot with a regression line.

It helps visualize the relationship (correlation) between two numerical variables and shows the best-fit line.

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Regression Plot of Total Bill vs Tip")
plt.show()
```

# 2. Residplot:
residplot() (Residual Plot) is used to visualize the residuals (errors) in a regression model.

A residual is the difference between the actual value and the predicted value.

Formula

Residual=Actual Value−Predicted Value

```python
sns.residplot(x="total_bill", y="tip", data=tips)
plt.title("Residual Plot")
plt.show()
```

# 3.Kdeplot:
A Bivariate KDE Plot is used to visualize the relationship and density distribution of two numerical variables.

Instead of showing individual points like a scatter plot, it shows density contours (similar to a topographic map) 
indicating where data points are concentrated.

```python
sns.kdeplot(x="total_bill", y="tip", fill=True, data=tips)
plt.title("Bivariate KDE Plot")
plt.xticks([5,10,15,20,25,30,35,40,45,50])
plt.yticks([0,2,4,6,8,10])
plt.show()
```

# PairGrid

```python
import seaborn as sns

iris = sns.load_dataset("iris")
```

```python
g=sns.PairGrid(iris,hue='species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
g.add_legend()
plt.show()
```

# Facet Grid

```python
tips = sns.load_dataset('tips')

g = sns.FacetGrid(tips, col='day', hue='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()

plt.show()
```

# Joint Grid

```python
g = sns.JointGrid(data = tips, x='total_bill', y='tip')
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.histplot)
plt.show()
```
