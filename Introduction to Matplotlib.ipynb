# MATPLOTLIB

Matplotlib in Python

Matplotlib is a Python library used for data visualization. It helps us create graphs, charts, and plots to understand 
data more easily.

Why Use Matplotlib?
   Visualize data clearly.
   Identify trends and patterns.
   Compare values.
   Create professional charts for reports and dashboards.
   Works well with NumPy and Pandas.

```python
#Import Matplotlib

import matplotlib.pyplot as plt

#pyplot is the most commonly used module in Matplotlib.
```

# Types of Data Analysis in Matplotlib

# 1. Univariate Analysis

```python
#Definition:
#  Analysis of one variable at a time to understand its distribution, spread, and central tendency.
#Example
#Analyze students' marks:

import matplotlib.pyplot as plt

marks = [45, 50, 55, 60, 65, 70, 75, 80]

plt.hist(marks)
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
#Common Plots
Histogram (plt.hist())
Box Plot (plt.boxplot())
Pie Chart (plt.pie())

#Purpose
Find distribution
Detect outliers
Understand data spread
```

# 2. Bivariate Analysis

```python
# Definition:
# Analysis of two variables to determine the relationship between them.
#Example
#Study hours vs marks:

import matplotlib.pyplot as plt

hours = [1, 2, 3, 4, 5]
marks = [40, 50, 60, 70, 80]

plt.scatter(hours, marks)
plt.title("Hours Studied vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

```python
#Common Plots
Scatter Plot (plt.scatter())
Line Plot (plt.plot())
Grouped Bar Chart

#Purpose
Find correlation
Compare two variables
Identify trends and relationships
```

# 3. Multivariate Analysis

```python
# Definition:

# Analysis of more than two variables simultaneously.
# Example
# Compare sales of multiple products across months:

import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr']
product_A = [100, 120, 140, 160]
product_B = [90, 110, 130, 150]

plt.plot(months, product_A, label='Product A')
plt.plot(months, product_B, label='Product B')

plt.title("Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
#Common Plots
Multiple Line Plots
Scatter Plots with color/size variations
Heatmaps (often using other libraries)
3D Plots

#Purpose
Analyze interactions among multiple variables
Discover patterns and dependencies
Support advanced business analysis
```

```python
#Quick Comparison

#Analysis Type	         Variables	         Example	              Common Plots
Univariate	                 1	         Marks Distribution	      Histogram, Box Plot
Bivariate	                 2	         Hours vs Marks	          Scatter Plot, Line Plot
Multivariate	          3 or More	     Month, Product, Sales	  Multiple Line Plots, Heatmaps
Interview Answer

Univariate Analysis: Studies one variable at a time to understand its distribution and characteristics.

Bivariate Analysis: Studies the relationship between two variables, such as correlation or trends.

Multivariate Analysis: Studies three or more variables together to understand complex relationships and patterns in the data.
```

# Types of Plots in Matplotlib

# 1. Line Plot

```python
#Used to show trends over time or continuous data.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.title("Line Plot")
plt.show()                                                    #Example: Monthly sales trend.
```

# 2. Scatter Plot

```python
#Used to show the relationship between two variables.

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()                                                          #Example: Study hours vs exam marks.
```

# 3. Bar Chart

```python
#Used to compare categories.

products = ['A', 'B', 'C']
sales = [100, 150, 120]

plt.bar(products, sales)
plt.title("Bar Chart")
plt.show()                                                          # Example: Sales by product.
```

# 4. Horizontal Bar Chart

```python
plt.barh(products, sales)
plt.title("Horizontal Bar Chart")
plt.show()                                                        #Example: Comparing employee performance.
```

# 5. Histogram

```python
#Shows the distribution of data.

data = [10, 20, 20, 30, 40, 40, 40, 50]

plt.hist(data)
plt.title("Histogram")
plt.show()                                                        #Example: Student marks distribution.
```

# 6. Pie Chart

```python
#Shows percentage contribution of categories.

sizes = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()                                                        #Example: Market share analysis.
```

# 7. Box Plot

```python
# Used to identify:

#Median
#Quartiles
#Outliers
data = [10, 12, 15, 18, 20, 25, 100]

plt.boxplot(data)
plt.show()                                                         #Example: Detecting outliers in salaries.
```

# 8. Area Plot

```python
#Shows cumulative totals over time.

x = [1,2,3,4]
y = [10,20,15,25]

plt.fill_between(x, y)
plt.show()                                                #Example: Revenue growth.
```

# 9. Stem Plot

```python
#Used for discrete data.

plt.stem([1,2,3,4], [10,20,15,25])
plt.show()                                                 #Example: Signal processing data.
```

# 10. Step Plot

```python
#Displays changes in steps.

plt.step([1,2,3,4], [10,20,15,25])
plt.show()                                                  #Example: Inventory levels.
```

# Subplot in Matplotlib

A subplot is used to display multiple plots in a single figure (window).

Instead of creating separate windows for each graph, subplots allow you to organize multiple graphs together for easy comparison.

```python
#Syntax
plt.subplot(nrows, ncols, index)

#Where:

nrows → Number of rows
ncols → Number of columns
index → Position of the plot (starts from 1)
```

```python
#Example 1: Two Plots in One Figure

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 35, 78, 50]
y2 = [5, 10, 18, 29, 25]

plt.subplot(1, 2, 1)   # 1 row, 2 columns, 1st plot
plt.plot(x, y1, color='blue')
plt.title("Plot 1")

plt.subplot(1, 2, 2)   # 1 row, 2 columns, 2nd plot
plt.plot(x, y2, color='green')
plt.title("Plot 2")

plt.tight_layout()   # Avoid overlap
plt.show()
```

Layout
+---------+---------+
| Plot 1  | Plot 2  |
+---------+---------+

```python
#Example 2: 2 Rows and 2 Columns

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 10, 15, 20, 25]
plt.plot(x, y1, color='blue', label = 'Even')
plt.plot(x, y2, color='green', label='odd')
plt.legend()
plt.show()
```

# Using Different Plot Types

```python
# import matplotlib.pyplot as plt

import numpy as np
fig, ax = plt.subplots(2, 2, figsize=(8, 6))

ax[0, 0].plot([1, 2, 3], [4, 5, 6])
ax[0, 1].bar(['A', 'B', 'C'], [5, 7, 3])
ax[1, 0].hist(np.random.randn(100))
ax[1, 1].scatter([1, 2, 3], [3, 2, 1])

fig.suptitle("Multiple Subplots Example")
plt.tight_layout()
plt.show()
#This displays a Line Plot and a Bar Plot side by side.
```

# Customized plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

plt.plot(x, y, color='red', marker='*', linestyle='--', linewidth=2, markersize=10)

plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True,linewidth=0.5, linestyle='dotted')
plt.xticks([1,2,3,4,5,6])
plt.yticks([10,20,30,40,50,60])
plt.show()
```
